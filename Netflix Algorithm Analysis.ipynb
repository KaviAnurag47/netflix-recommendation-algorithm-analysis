- Duration Info

````python
def categorize_duration(duration):
    if pd.isna(duration):
        return 'Unknown'
    duration = str(duration)
    if 'Season' in duration:
        num = int(duration.split()[0])
        if num == 1:
            return '1 Season'
        elif num == 2:
            return '2 Seasons'
        else:
            return '3+ Seasons'
    elif 'min' in duration:
        num = int(duration.split()[0])
        if num < 60:
            return '< 60 min'
        elif num <= 90:
            return '60–90 min'
        elif num <= 120:
            return '91–120 min'
        else:
            return '> 120 min'
    else:
        return 'Other'

# Apply the grouping
df['Duration Category'] = df['Duration Info'].apply(categorize_duration)

# Count categories
duration_counts = df['Duration Category'].value_counts().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8, 5))
ax = sns.barplot(x=duration_counts.index,
                 y=duration_counts.values,
                 palette='Set2',
                 edgecolor='black')

# Add annotations
for i, v in enumerate(duration_counts.values):
    ax.text(i, v + 20, str(v), ha='center', fontsize=10)

# Style
plt.title('Grouped Duration Categories', fontsize=14, pad=20)
plt.ylabel('Count')
plt.xlabel('')
plt.xticks(rotation=15)
sns.despine()
plt.tight_layout()
plt.show()
````
